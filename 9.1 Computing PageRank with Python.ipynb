# Computing the page rank of a graph

Its incredible easy to compute the Page rank of any graph using Python.

Firt, let's import a module that will allow us to do that.
This module is called "networkx", and is a Python module that has a lot of pre-coded graph algorithms 

```python
import networkx as nx
from operator import itemgetter
import matplotlib.pyplot as plt 
import numpy as np
```

## 1. A simple Example

The next step is to create a graph stored in a file. Download the file from http://bit.ly/aSimpleGraph

Our file has the information stored as follows:
- every line has a website that points to another website
- every line is \t separated
- the first item for every entry is the website that is pointing, and the second item is the website pointed to
- the third item captures number of links information

```python
# create an empty graph
G=nx.DiGraph()

# read your data line by line and start adding nodes and edges to the graph
with open('flow.txt', mode='r', encoding='utf-8') as f:
    for line in f:
        # read information of which website (source) points to which website (target), and how many times (w)
        source,target,w=line.strip().split('\t')
        # add weighted edges to graph
        w=int(w)
        G.add_edge(source,target,weight=w)
```

Computing the page rank of the graph is now a piece of cake.

```python
pr = nx.pagerank(G, alpha=0.85,weight='weight')

srt = sorted(pr.items(),key=itemgetter(1),reverse=True) 

srt[:50]
```

# 2. Another example

Let's now go and create some graphs step by step

```python
G = nx.DiGraph()
G.add_nodes_from("ABCD")
nx.draw(G, with_labels=True)
plt.show()
```

What is the pagerank of this graph?

```python
pr=nx.pagerank(G, alpha=0.1)
pr
```

Let's do another example

```python
G = nx.DiGraph()
G.add_nodes_from([1,2,3,4])
G.add_weighted_edges_from([(1,3,1), (1,4, 1),(1,2,1),(2,3,1),(2,4,1),(3,1,1),(4,1,1),(4,3,1)])
nx.draw(G, with_labels=True)
plt.show()
```

```python
# compute pagerank
pr=nx.pagerank(G,alpha=0)
pr
```
